# Análise de Evasão de Clientes - TelecomX BR

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/Lidianecrsantos/TelecomX_BR_eda/refs/heads/main/telecomx_processed_data.csv'
df = pd.read_csv(url)
display(df.head())
```

```python

df = df.drop('customerID', axis=1)


display(df.head())
print(df.columns)
```

```python

categorical_cols = df.select_dtypes(include=['object']).columns

if 'Churn' in categorical_cols:
    categorical_cols = categorical_cols.drop('Churn')



df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)


display(df.head())
print(df.columns)
```

```python

churn_counts = df['Churn'].value_counts()


churn_proportions = df['Churn'].value_counts(normalize=True)


print("Contagem de clientes por classe (0: Ativo, 1: Evadido):")
print(churn_counts)
print("\nProporção de clientes por classe (0: Ativo, 1: Evadido):")
print(churn_proportions)


if churn_proportions.min() < 0.4: # Limiar comum para considerar desequilíbrio
    print("\nAtenção: Há um desequilíbrio significativo nas classes de 'Churn'.")
else:
    print("\nNão há um desequilíbrio significativo nas classes de 'Churn'.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


correlation_matrix = df.corr()


plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação das Variáveis')
plt.show()


print("\nCorrelação com a variável 'Churn':")
print(correlation_matrix['Churn'].sort_values(ascending=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(8, 6))
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title('Tempo de Contrato vs. Evasão')
plt.xlabel('Evasão (0: Não, 1: Sim)')
plt.ylabel('Tempo de Contrato (meses)')
plt.xticks([0, 1], ['Não Evadiu', 'Evadiu'])
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x='Churn', y='Charges.Total', data=df)
plt.title('Total Gasto vs. Evasão')
plt.xlabel('Evasão (0: Não, 1: Sim)')
plt.ylabel('Total Gasto')
plt.xticks([0, 1], ['Não Evadiu', 'Evadiu'])
plt.show()
```

Isso também sugere overfitting, embora ligeiramente menos severo que a Árvore de Decisão individual, pois Random Forest é um ensemble que reduz um pouco o overfitting de árvores individuais.





Em resumo, a **Regressão Logística** teve o melhor desempenho no conjunto de teste e não mostrou sinais claros de overfitting severo. Os modelos baseados em árvore (Decision Tree e Random Forest) e o KNN apresentaram overfitting, indicando a necessidade de ajustar seus hiperparâmetros ou considerar outras técnicas de regularização.

    *   **Tempo de Contrato (`tenure`):** O boxplot de `tenure` vs. `Churn` demonstrou claramente que clientes que evadiram (`Churn` = 1) geralmente possuem um tempo de contrato significativamente menor do que aqueles que não evadiram (`Churn` = 0).
    *   **Total Gasto (`Charges.Total`):** O boxplot de `Charges.Total` vs. `Churn` indicou que clientes que evadiram tendem a ter um gasto total acumulado menor em comparação com os clientes que permanecem.

## 5. Análise de Fatores de Evasão

Com base nos coeficientes da Regressão Logística otimizada, na importância das variáveis do Random Forest otimizado e nas visualizações exploratórias, os principais fatores que impulsionam a evasão de clientes foram identificados:

*   **Tempo de Contrato (`tenure`):** Clientes com menor tempo de permanência na empresa são significativamente mais propensos a evadir. Este é um dos preditores mais fortes de evasão.
*   **Tipo de Contrato:** Clientes com contratos mensais (`Contract_Month-to-month`) têm uma probabilidade muito maior de evadir em comparação com aqueles com contratos de um ou dois anos (`Contract_One year`, `Contract_Two year`).
*   **Serviços de Internet:** Ter serviço de internet, especialmente fibra óptica (`InternetService_Fiber optic`), está associado a uma maior probabilidade de evasão (possivelmente devido a custos mais altos ou problemas de serviço), enquanto não ter serviço de internet (`InternetService_No`) está fortemente associado à permanência (estes clientes podem ter menos opções ou necessidades).
*   **Serviços Online Adicionais:** A ausência destes serviços está associada a uma maior probabilidade de evasão, enquanto a assinatura destes serviços está associada à permanência.
*   **Método de Pagamento:** O uso de cheque eletrônico (`PaymentMethod_Electronic check`) é um forte indicador de evasão. Métodos de pagamento automáticos (transferência bancária, cartão de crédito) e cheque enviado por correio estão associados à permanência.
*   **Gasto Total (`Charges.Total`) e Gasto Mensal (`Charges.Monthly`)/Contas Diárias (`Contas_Diarias`):** Clientes com menor gasto total acumulado e maior gasto mensal tendem a evadir.
*   **Faturamento Sem Papel (`PaperlessBilling`):** Clientes que optam por faturamento sem papel são ligeiramente mais propensos a evadir.
*   **Parceiro (`Partner`) e Dependentes (`Dependents`):** Clientes sem parceiro ou dependentes são um pouco mais propensos a evadir.
*   **Senior Citizen (`SeniorCitizen`):** Clientes idosos (SeniorCitizen=1) são ligeiramente mais propensos a evadir.

**Resumo dos Principais Fatores:** A evasão parece estar concentrada em clientes mais novos, com contratos flexíveis (mês a mês), que utilizam serviços de internet de alta velocidade (fibra óptica) mas não assinam serviços de segurança/suporte adicionais, que utilizam métodos de pagamento menos "estáveis" como o cheque eletrônico, e que possuem um histórico de gasto total menor.


## 6. Estratégias de Retenção

Com base nos principais fatores de evasão identificados (Tempo de Contrato, Tipo de Contrato, Serviços de Internet e Adicionais, Método de Pagamento, Gasto Total/Mensal, etc.), as seguintes estratégias de retenção são propostas:

*   **Clientes com Baixo Tempo de Contrato (`tenure`):**
    *   **Estratégia:** Implementar um programa de boas-vindas e fidelidade intensivo nos primeiros meses. Oferecer benefícios exclusivos, suporte proativo e acompanhamento personalizado para garantir uma experiência positiva inicial. Considerar ofertas de desconto ou upgrades após um curto período (ex: 3 ou 6 meses) para incentivar a permanência.
    *   **Justificativa:** Clientes novos são mais propensos a evadir. Uma experiência positiva e incentivos iniciais podem construir lealdade e superar possíveis dificuldades de adaptação.

*   **Clientes com Contratos Mensais (`Contract_Month-to-month`):**
    *   **Estratégia:** Oferecer incentivos atraentes (descontos significativos na mensalidade, meses grátis, upgrades de serviço) para que migrem para contratos de longo prazo (um ou dois anos). Comunicar claramente os benefícios e a economia a longo prazo de contratos mais longos.
    *   **Justificativa:** Contratos mensais oferecem flexibilidade para o cliente, mas representam alto risco de evasão para a empresa. Incentivar a mudança para contratos mais longos aumenta a estabilidade da base de clientes.

*   **Clientes com Serviço de Fibra Óptica (`InternetService_Fiber optic`) sem Serviços Adicionais de Segurança/Suporte (`OnlineSecurity_No`, `TechSupport_No`, etc.):**
    *   **Estratégia:** Segmentar estes clientes e oferecer pacotes ou descontos especiais para a adição de serviços de segurança online, suporte técnico, backup, etc. Destacar o valor e a tranquilidade que estes serviços proporcionam, especialmente para usuários de alta velocidade.
    *   **Justificativa:** Clientes de fibra óptica que evadem podem estar enfrentando problemas de segurança, suporte ou simplesmente não veem valor nos serviços básicos. Oferecer soluções integradas pode aumentar a satisfação e a percepção de valor.

*   **Clientes Utilizando Cheque Eletrônico (`PaymentMethod_Electronic check`):**
    *   **Estratégia:** Incentivar a migração para métodos de pagamento automáticos (transferência bancária, cartão de crédito) oferecendo pequenos descontos ou benefícios. Investigar proativamente se há problemas ou insatisfação associados ao método de pagamento via cheque eletrônico e tentar resolver as causas raiz.
    *   **Justificativa:** O cheque eletrônico está associado a maior evasão. Métodos automáticos podem indicar maior compromisso com o serviço e reduzir atritos no processo de pagamento.

*   **Clientes com Baixo Gasto Total (`Charges.Total`) e/ou Alto Gasto Mensal (`Charges.Monthly`)/Contas Diárias (`Contas_Diarias`):**
    *   **Estratégia:** Analisar o padrão de uso destes clientes. Se o gasto total é baixo, pode indicar subutilização ou planos inadequados; oferecer consultoria para otimizar o plano ou pacotes mais adequados. Se o gasto mensal é alto em relação ao tempo de contrato/serviços, pode indicar insatisfação com o custo-benefício; considerar ofertas personalizadas ou renegociação.
    *   **Justificativa:** O padrão de gasto reflete a percepção de valor. Intervenções direcionadas podem alinhar o serviço ao valor percebido pelo cliente.

*   **Clientes com Faturamento Sem Papel (`PaperlessBilling`):**
    *   **Estratégia:** Embora associado à evasão, este fator pode estar ligado a clientes mais digitalmente engajados. A estratégia deve focar em manter o engajamento através de canais digitais, oferecendo conteúdo relevante, suporte online eficiente e promoções exclusivas via e-mail/app.
    *   **Justificativa:** Entender o perfil digital destes clientes permite uma comunicação mais eficaz e personalizada.

*   **Clientes Sem Parceiro (`Partner_No`) ou Dependentes (`Dependents_No`) e Senior Citizens (`SeniorCitizen`):**
    *   **Estratégia:** Adaptar a comunicação e as ofertas a estes segmentos. Para Senior Citizens, garantir canais de suporte acessíveis e compreensíveis. Para clientes sem parceiro/dependentes, focar em ofertas individuais ou benefícios que não dependam de múltiplos usuários.
    *   **Justificativa:** Segmentos demográficos específicos podem ter necessidades e preferências distintas que influenciam a evasão.

**Considerações Gerais para Implementação:**

*   **Segmentação:** As estratégias devem ser direcionadas aos segmentos de clientes em maior risco, conforme identificado pelos modelos preditivos.
*   **Comunicação Proativa:** Utilizar os modelos para identificar clientes em risco e contatá-los proativamente com ofertas de retenção personalizadas antes que decidam evadir.
*   **Feedback Contínuo:** Estabelecer canais para coletar feedback dos clientes e usar essas informações para refinar as estratégias e melhorar a qualidade do serviço.
*   **Monitoramento:** Monitorar continuamente a eficácia das estratégias de retenção através de métricas de redução da taxa de evasão nos segmentos alvo.

Estas estratégias visam abordar os principais motivos de evasão e aumentar a probabilidade de retenção dos clientes da TelecomX.


## 7. Conclusão

A análise dos dados e a modelagem preditiva foram bem-sucedidas na identificação dos principais fatores que contribuem para a evasão de clientes na TelecomX. A Regressão Logística otimizada demonstrou ser o modelo mais eficaz para prever a evasão, apresentando o melhor equilíbrio entre Precision e Recall no conjunto de teste desbalanceado. Os insights sobre o tempo de contrato, tipo de contrato, serviços utilizados, métodos de pagamento e histórico de gastos fornecem uma base sólida para a TelecomX desenvolver e implementar estratégias de retenção direcionadas, visando reduzir a taxa de evasão e aumentar a satisfação e o lifetime value dos clientes.
